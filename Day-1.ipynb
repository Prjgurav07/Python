1. Write a program to do arithmetical operation

#### Input is used to get input from the user. input enter by the user, system read it as string value. so to convert that value in integer or decimal we use int or float

```python
num1 = int(input('Enter a number 1:'))
num2 = int(input('Enter a number 2:'))
add = num1 + num2
sub = num1 - num2
multiply = num1 * num2
division = num1/num2
print(add)
print(sub)
print(multiply)
print(division)
```

##### 2. Calculator

```python
n1 = int(input('Enter 1st number:'))
n2 = int(input('Enter 2nd number:'))
op = input('Enter operation')
add = n1+n2
sub = n1-n2
multiple = n1 * n2
divide = n1/n2
if op == "+":
    print(add)
elif op == "-":
    print(sub)
elif op == "*":
    print(multiple)
elif op == "/":
    print(divide)
else:
    print('Enter valid operator')
```
