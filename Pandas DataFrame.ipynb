### Pandas DataFrame

#### Create DataFrame

```python
import pandas as pd
```

```python
data = pd.DataFrame({'Name':['Anna','Ben'],
                    'Lemonade':[2,3]})
```

```python
data
```

```python
s1 = pd.Series(data=[1,2,3,4])
```

```python
s2 = pd.Series(data=[10,20,30,40])
```

```python
data2 = pd.DataFrame({'C1':s1,'C2':s2})
```

```python
data2
```

```python
data2.to_csv('lemonade.csv')
```

```python
df = pd.read_csv('lemonade.csv')
```

```python
df
```

```python
import numpy as np
```

```python
dictt = {'x':[1,2,3],'y':np.array([10,20,30]),'z':50}
```

```python
listt = [[1,2,3,4],[10,20,30,40],[100,200,300,400]]
```

```python
pd.DataFrame(listt,columns=['x','y','z','a'])
```

```python
narr = np.array([[1,2,100],[2,4,100],[3,6,100]])
```

```python
pd.DataFrame(narr,columns=['x','y','z'])
```
