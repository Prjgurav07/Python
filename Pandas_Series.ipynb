### Pandas Series

##### Pandas is used to data manuplation and analysis. In this we can handle missing data, outlies and analysis. Pandas Series is 1D table in which we have only 1 column.

```python
import pandas as pd
```

```python
sales = pd.Series([200,679,346,765,750,280,543,120,450,654,320,576,876,980])
```

```python
sales
```

#### Indexing

##### By using index we can find value at particular index(location)

```python
sales[7]
```

```python
sales[3]
```

#### Slicing

##### Slicing is used to fetch particular range. In this last index is not include. 

```python
sales[2:5]
```

```python
sales[5:]
```

#### Attributes

##### Index function is used fetch index.

```python
sales.index
```

##### Values function fetch values of Series

```python
sales.values
```

##### dtypes is used to tell data type of Series values

```python
sales.dtypes
```

##### head() is used to fetch top 5 rows.

```python
sales.head()
```

##### tail() is used to fetch bottom 5 rows

```python
sales.tail()
```

##### sort_values() is used to sort values in ascending order.

```python
sales.sort_values()
```

##### sort_values(ascending=False) is used to sort values in descending order

```python
sales.sort_values(ascending=False)
```

##### count() is used to count number of rows in Series

```python
sales.count()
```

##### isnull() is used to check values is null or not

```python
sales.isnull()
```

##### Operation is used to do operation (add, multiple, sub, power)  to values nu constant

```python
sales + 100
```

```python
sales ** 2
```

### Handle Missing

```python
sales.isnull
```

##### fillna is used to fill missing values with constant or mea

```python
sales.fillna(0)
```
